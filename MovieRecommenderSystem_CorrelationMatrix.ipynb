```python
import pandas as pd 
```

```python
rating_cols=['User Id','Movie_ID','Rating']
ratings_db=pd.read_csv("C:\\Users\\91839\\MLCourse\\ml-100k\\u.data",sep='\t',
                       names=rating_cols,usecols=range(3),encoding="ISO-8859-1")
ratings_db.head()
```

```python
movie_cols=['Movie_ID','Movie_title']
movie_db=pd.read_csv("C:\\Users\\91839\\MLCourse\\ml-100k\\u.item",sep='|',
                     names=movie_cols,usecols=range(2),encoding="ISO-8859-1")
movie_db.head()
```

```python
ratings_db=pd.merge(ratings_db,movie_db)
ratings_db.head()
```

```python
user_ratings = ratings_db.pivot_table(index=['User Id'],columns=['Movie_title'],
                                   values='Rating')
user_ratings.head()
```

```python
corrMatrix=user_ratings.corr(method='pearson',min_periods=50)
# we need to incorporate min_periods so as to include popular movies in recommendations 
corrMatrix.head()
```

```python
user1=user_ratings.loc[0].dropna()
'''
we took a sample user from the sparsematrix and created a dictionary object 
with movie title and user ratings. Novie title is the index variable 
'''
```

```python
sim_movies=pd.Series([])
#initializing the series object which will contain the similar movies
for i in range(len(user1.index)):
    print ("Adding Movie Recommendations for "+ user1.index[i])
    movies = corrMatrix[user1.index[i]].dropna()
    #extracting the movies list similarity score for each movie rated by user 
    movies = movies.map(lambda x: x * user1[i])
    #multiplying movie rating given by user with similarity score because movie recommendation is based on s
    #similarity as well as choice. rating shows choice. hence a recommendation score is required
    sim_movies = pd.concat([sim_movies,pd.Series(movies)])

print ("Sorting..")
sim_movies = sim_movies.groupby(sim_movies.index).sum()
#some movies may be recommended more times if they bear similarity with more than
# one movie which is seen by sample user. hence we take the sum of recommendation score 
# of each movie. 
sim_movies.sort_values(inplace = True,ascending = False)
print (sim_movies.head(10))
```

```python
filtered_sim_movies= sim_movies.drop(user1.index)
filtered_sim_movies.head()
```
